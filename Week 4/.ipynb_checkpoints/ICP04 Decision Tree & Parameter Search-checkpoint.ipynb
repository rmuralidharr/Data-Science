## Import required packages

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt

from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV


from sklearn.tree import DecisionTreeClassifier 
from sklearn.svm import SVC

from sklearn.metrics import confusion_matrix


np.random.seed(1)
```

## load data 

```python
X_train =pd.read_csv(r'~/Murali/DSP/Week 4/airbnb_train_X_price_gte_150.csv')
y_train = pd.read_csv(r'~/Murali/DSP/Week 4/airbnb_train_y_price_gte_150.csv') 
```

```python
X_test = pd.read_csv("airbnb_test_X_price_gte_150.csv") 
y_test = pd.read_csv("airbnb_test_y_price_gte_150.csv")
```

## perform tab

```python
performance = pd.DataFrame({"model": [], "Accuracy": [], "Precision": [], "Recall": [], "F1": []})
```

## SVM

```python
%%time
param_grid = {'C': [0.1, 1, 10, 50], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'degree':[1,2,3,4],
              'kernel': ['poly']} 
  
SVC_grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
SVC_grid.fit(X_train,np.ravel(y_train))

print(SVC_grid.best_params_)
print(SVC_grid.best_estimator_)
```

```python
model_preds = SVC_grid.predict(X_test)
print(len(model_preds))
c_matrix = confusion_matrix(y_test, model_preds)
TP = c_matrix[1][1]
TN = c_matrix[0][0]
FP = c_matrix[0][1]
FN = c_matrix[1][0]
performance = pd.concat([performance, pd.DataFrame({'model':"poly svm", 
                                                    'Accuracy': [(TP+TN)/(TP+TN+FP+FN)], 
                                                    'Precision': [TP/(TP+FP)], 
                                                    'Recall': [TP/(TP+FN)], 
                                                    'F1': [2*TP/(2*TP+FP+FN)]
                                                     }, index=[0])])
```

# Decision Tree

```python
%%time
score_measure = "accuracy"
kfolds = 5

param_grid = {
    'min_samples_split': np.arange(1,50),  
    'min_samples_leaf': np.arange(1,50),
    'min_impurity_decrease': np.arange(0.0001, 0.01, 0.0005),
    'max_leaf_nodes': np.arange(5, 70), 
    'max_depth': np.arange(1,50), 
    'criterion': ['entropy', 'gini'],
}

dtree = DecisionTreeClassifier()
rand_search = RandomizedSearchCV(estimator = dtree, param_distributions=param_grid, cv=kfolds, n_iter=750,
                           scoring=score_measure, verbose=1, n_jobs=-1,  # n_jobs=-1 will utilize all available CPUs 
                           return_train_score=True)

_ = rand_search.fit(X_train,y_train)

print(f"The best {score_measure} score is {rand_search.best_score_}")
print(f"... with parameters: {rand_search.best_params_}")

bestRecallTree = rand_search.best_estimator_
```

```python
c_matrix = confusion_matrix(y_test, rand_search.predict(X_test))
TP = c_matrix[1][1]
TN = c_matrix[0][0]
FP = c_matrix[0][1]
FN = c_matrix[1][0]
print(f"Accuracy={(TP+TN)/(TP+TN+FP+FN):.7f} Precision={TP/(TP+FP):.7f} Recall={TP/(TP+FN):.7f} F1={2*TP/(2*TP+FP+FN):.7f}")
performance = pd.concat([performance, pd.DataFrame({'model':"trees", 
                                                    'Accuracy': [(TP+TN)/(TP+TN+FP+FN)], 
                                                    'Precision': [TP/(TP+FP)], 
                                                    'Recall': [TP/(TP+FN)], 
                                                    'F1': [2*TP/(2*TP+FP+FN)]
                                                     }, index=[0])])
```
