# Predicting the 10th day closing stock price by using past 9 days data. For this assignment i have taken 100 days data for PAPA JOHNS company.
# I would first do data processing and then apply RNN, LTSM, Gru, and Conv1D  models and observe the performance of the results.

# Import all the rquired labraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

np.random.seed(1)
tf.random.set_seed(1)
```

# Load the data

```python
stocks=pd.read_csv("PZZA.csv")
```

```python
stocks.columns
```

```python
stocks['Close'].isna().sum()
```

```python
df = stocks[['Close']] # watch the spaces in the column name 
```

```python
row_count = df.shape[0] # store the number of obersvations (daily temperature readings)
row_count
```

```python
stock = df.to_numpy() # to make things easier, we create a numpy array from our dataframe of temperatures
stock.shape
```

```python
sequence_length = 10  # here, I create a variable to hold the number of days that will be in a sequence. This makes it easier to change later.
```

```python
stock = stock[row_count % sequence_length: ] # Since our data may not be a multiple of sequence_length, we will ignore 
# the first number of days in order to make the total a number divisible by sequence length
stock.shape
```

```python
stock = stock.reshape(row_count // sequence_length, sequence_length) # note that temps are a multiple of sequence length, we can split the data into rows of sequence length.
stock
```

```python
# we can now see that our days have been organized into rows of sequence length. 
```

```python
# create our X and y. 
# X will be the 9 days prior...
X = stock[:, :-1] # all rows, all columns except the last one
# y will be the 10th day (what we are trying to predict)
y = stock[:, -1] # all rows, only the last column
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

print(X_train.shape)
print(X_test.shape)
```

 # Simple RNN with one layer

```python
print(X_train.shape)  # 2d array of shape (n_samples, n_features)
print(X_test.shape)  # 1d array of shape (n_samples,)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, GRU, Conv1D

n_steps = sequence_length - 1
n_inputs = 1

model = Sequential(
    [
        SimpleRNN(64, input_shape=[n_steps, n_inputs]),
        Dense(1, activation=None)
    ]
)
model.summary()
```

```python
# If multiclass, use "sparse_categorical_crossentropy" as the loss function
model.compile(
    loss="mean_squared_error", 
    optimizer=tf.keras.optimizers.Nadam(learning_rate=0.001), 
    metrics=[tf.keras.metrics.RootMeanSquaredError()]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

import time
start_time = time.time()

early_stop = EarlyStopping(monitor='val_root_mean_squared_error', min_delta = 0.0, patience=10, verbose=1, mode='min')

history = model.fit(
    X_train, 
    y_train, 
    epochs=600,
    validation_data=(X_test, y_test), 
    callbacks=[early_stop]
)

end_time = time.time()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
fig.suptitle('Simple RNN Model Evaluation')

ax1.set_title('loss')
#ax1.set_ylim(0, 100)
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.legend(['train', 'test'], loc='upper left')

ax2.set_title('RMSE')
#ax2.set_ylim(0, 1)
ax2.plot(history.history['root_mean_squared_error'])
ax2.plot(history.history['val_root_mean_squared_error'])
ax2.legend(['train', 'test'], loc='upper left')
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)  # these predictions are probabilities (0-1) taken from the sigmoid function

mean_squared_error(y_test, y_pred)**0.5
```

```python
prior_days = np.array(
    [
        77,75,79,85,78,75,81,83,87,
    ]
)

# reshape into the three dimensions for input into model
prior_days = prior_days.reshape(1, 9, 1) # 1 row, 9 time steps (features), and 1 output/prediction)
```

```python
model.predict(prior_days) # note the structure of the prediction
```

```python
tomorrows_stock_simplernn = model.predict(prior_days)[0][0] # we see the predicted temp is embedded inside two dimensions
```

```python
print(f"Tenth day closing stock price is  {tomorrows_stock_simplernn:.2f} dollars.")
```

```python
from tensorflow.keras.callbacks import EarlyStopping

import time
start_time = time.time()

early_stop = EarlyStopping(monitor='val_root_mean_squared_error', min_delta = 0.0, patience=10, verbose=1, mode='min')

history1 = model.fit(
    X_train, 
    y_train, 
    epochs=600,
    validation_data=(X_test, y_test), 
    callbacks=[early_stop]
)

end_time = time.time()
```
