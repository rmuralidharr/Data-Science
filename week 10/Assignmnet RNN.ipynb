# Import required libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

np.random.seed(1)
tf.random.set_seed(1)
```

# Load the data

```python
dell_stocks=pd.read_csv(r"~/Murali/DSP/week 10/DELL.csv")
```

```python
dell_stocks.head()
```

```python
dell_stocks.isna().sum()
```

```python
# store original data into operation variable for data handling
dell_st=dell_stocks[:100]
```

```python
dell_st = dell_st[['Close']] # watch the spaces in the column name 
```

```python
rowCount = dell_st.shape[0] # store the number of obersvations 
rowCount
```

```python
stock = dell_st.to_numpy() # to make things easier, we create a numpy array from our dataframe 
stock.shape
```

```python
sequence_length = 10  # In this case, I establish a variable to contain the number of days in a series. This makes further changes easy.
```

```python
stock = stock[rowCount % sequence_length: ]
stock.shape
```

```python
stock = stock.reshape(rowCount // sequence_length, sequence_length) 
stock
```

###  We can now see that our days are structured into rows of increasing length.

```python
X = stock[:, :-1] #  all columns,all rows, except the last one
y = stock[:, -1] # all rows, only the last column
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train.shape
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

print(X_train.shape)
print(X_test.shape)
```

# RNN

```python
model = keras.models.Sequential([
    keras.layers.SimpleRNN(32, activation='relu', return_sequences=True, input_shape=[9, 1]),
    keras.layers.SimpleRNN(32, activation='relu', return_sequences=False),
    keras.layers.Dense(1, activation=None)
])
```

```python
# If multiclass, use "sparse_categorical_crossentropy" as the loss function
model.compile(
    loss="mean_squared_error", 
    optimizer=tf.keras.optimizers.Nadam(learning_rate=0.001), 
    metrics=[tf.keras.metrics.RootMeanSquaredError()]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

import time
start_time = time.time()

early_stop = EarlyStopping(monitor='val_root_mean_squared_error', min_delta = 0.0, patience=10, verbose=1, mode='min')

history_RNN = model.fit(
    X_train, 
    y_train, 
    epochs=600,
    validation_data=(X_test, y_test), 
    callbacks=[early_stop]
)

end_time = time.time()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
fig.suptitle('RNN Model Evaluation')

ax1.set_title('loss')
#ax1.set_ylim(0, 100)
ax1.plot(history_RNN.history['loss'])
ax1.plot(history_RNN.history['val_loss'])
ax1.legend(['train', 'test'], loc='upper left')

ax2.set_title('RMSE')
#ax2.set_ylim(0, 1)
ax2.plot(history_RNN.history['root_mean_squared_error'])
ax2.plot(history_RNN.history['val_root_mean_squared_error'])
ax2.legend(['train', 'test'], loc='upper left')
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)  # these predictions are probabilities (0-1) taken from the sigmoid function

mean_squared_error(y_test, y_pred)**0.5
```

```python
prior_days = np.array(
    [
        43,41,38,39,34,32,30,34,38,
    ]
)

prior_days = prior_days.reshape(1, 9, 1) 
```

```python
model.predict(prior_days) # note the structure of the prediction
```

```python
tomorrows_stock_deepRNN = model.predict(prior_days)[0][0] # we see the predicted temp is embedded inside two dimensions
```

```python
print(f"10th day closing stock price is  {tomorrows_stock_deepRNN:.2f} dollars.")
```

# LSTM layers

```python
model = keras.models.Sequential([
    keras.layers.LSTM(32, activation='tanh', return_sequences=True, input_shape=[9, 1]),
    keras.layers.LSTM(32, activation='tanh', return_sequences=False),
    keras.layers.Dense(1, activation=None)
])
```

```python
# If multiclass, use "sparse_categorical_crossentropy" as the loss function
model.compile(
    loss="mean_squared_error", 
    optimizer=tf.keras.optimizers.Nadam(learning_rate=0.001), 
    metrics=[tf.keras.metrics.RootMeanSquaredError()]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

import time
start_time = time.time()

early_stop = EarlyStopping(monitor='val_root_mean_squared_error', min_delta = 0.0, patience=10, verbose=1, mode='min')

history_LSTM = model.fit(
    X_train, 
    y_train, 
    epochs=600,
    validation_data=(X_test, y_test), 
    callbacks=[early_stop]
)

end_time = time.time()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
fig.suptitle('LSTM with more layers  Model Evaluation')

ax1.set_title('loss')
#ax1.set_ylim(0, 100)
ax1.plot(history_LSTM.history['loss'])
ax1.plot(history_LSTM.history['val_loss'])
ax1.legend(['train', 'test'], loc='upper left')

ax2.set_title('RMSE')
#ax2.set_ylim(0, 1)
ax2.plot(history_LSTM.history['root_mean_squared_error'])
ax2.plot(history_LSTM.history['val_root_mean_squared_error'])
ax2.legend(['train', 'test'], loc='upper left')
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)  # these predictions are probabilities (0-1) taken from the sigmoid function

mean_squared_error(y_test, y_pred)**0.5
```

# GRU layers

```python
model = keras.models.Sequential([
    keras.layers.GRU(32, activation='relu', return_sequences=True, input_shape=[9, 1]),
    keras.layers.GRU(32, activation='relu', return_sequences=False),
    keras.layers.Dense(1, activation=None)
])
```

```python

model.compile(
    loss="mean_squared_error", 
    optimizer=tf.keras.optimizers.Nadam(learning_rate=0.001), 
    metrics=[tf.keras.metrics.RootMeanSquaredError()]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

import time
start_time = time.time()

early_stop = EarlyStopping(monitor='val_root_mean_squared_error', min_delta = 0.0, patience=10, verbose=1, mode='min')

history_GRU = model.fit(
    X_train, 
    y_train, 
    epochs=600,
    validation_data=(X_test, y_test), 
    callbacks=[early_stop]
)

end_time = time.time()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
fig.suptitle('GRU with more layers  Model Evaluation')

ax1.set_title('loss')
#ax1.set_ylim(0, 100)
ax1.plot(history_GRU.history['loss'])
ax1.plot(history_GRU.history['val_loss'])
ax1.legend(['train', 'test'], loc='upper left')

ax2.set_title('RMSE')
#ax2.set_ylim(0, 1)
ax2.plot(history_GRU.history['root_mean_squared_error'])
ax2.plot(history_GRU.history['val_root_mean_squared_error'])
ax2.legend(['train', 'test'], loc='upper left')
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)  

mean_squared_error(y_test, y_pred)**0.5
```

# Conv1D

```python
model = keras.models.Sequential([
    keras.layers.Conv1D(filters=20, kernel_size=3, strides=1, padding="valid", input_shape=[9, 1]),
    keras.layers.GRU(32, activation='relu', return_sequences=True),
    keras.layers.GRU(32, activation='relu', return_sequences=False),
    keras.layers.Dense(1, activation=None)
])
```

```python

model.compile(
    loss="mean_squared_error", 
    optimizer=tf.keras.optimizers.Nadam(learning_rate=0.001), 
    metrics=[tf.keras.metrics.RootMeanSquaredError()]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

import time
start_time = time.time()

early_stop = EarlyStopping(monitor='val_root_mean_squared_error', min_delta = 0.0, patience=10, verbose=1, mode='min')

history_Conv1D = model.fit(
    X_train, 
    y_train, 
    epochs=600,
    validation_data=(X_test, y_test), 
    callbacks=[early_stop]
)

end_time = time.time()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
fig.suptitle('CONV 1D Model Evaluation')

ax1.set_title('loss')
#ax1.set_ylim(0, 100)
ax1.plot(history_Conv1D.history['loss'])
ax1.plot(history_Conv1D.history['val_loss'])
ax1.legend(['train', 'test'], loc='upper left')

ax2.set_title('RMSE')
#ax2.set_ylim(0, 1)
ax2.plot(history_Conv1D.history['root_mean_squared_error'])
ax2.plot(history_Conv1D.history['val_root_mean_squared_error'])
ax2.legend(['train', 'test'], loc='upper left')
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)  

mean_squared_error(y_test, y_pred)**0.5
```
